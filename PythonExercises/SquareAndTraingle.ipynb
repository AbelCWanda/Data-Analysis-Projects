```python
def desenhar_quadrado(tamanho):
    """Desenha um quadrado com o caractere #"""
    print("QUADRADO:")
    for i in range(tamanho):
        for j in range(tamanho):
            print("#", end=" ")
        print()  # Nova linha após cada linha do quadrado
    print()  # Espaço após o quadrado

def desenhar_triangulo(tamanho):
    """Desenha um triângulo com o caractere #"""
    print("TRIÂNGULO:")
    for i in range(1, tamanho + 1):
        # Imprime espaços em branco para centralizar
        espacos = tamanho - i
        print(" " * espacos, end="")
        
        # Imprime os caracteres # para formar o triângulo
        print("# " * i)
    print()  # Espaço após o triângulo

# Tamanho das figuras (pode ser ajustado)
tamanho_figuras = 36

# Desenhar as figuras
desenhar_quadrado(tamanho_figuras)
desenhar_triangulo(tamanho_figuras)
```
