### Calc Distance elevator

```python

tot_dist = 0
list_sequence = []

dist_elevator = int(input("insert the distance between floors [m]: "))
list_sequence.append(dist_elevator)

tot_travel = int(input("How many travels do you want: "))
list_sequence.append(tot_travel)

floor1 = int(input("Insert the floot where you are: "))
list_sequence.append(floor1)

while (tot_travel > 0):
    
    print(f"you are at {floor1}º floor")
    floor2 = int(input("Where do you want to go: "))

    dif12 = floor2 - floor1
    if dif12 < 0:
        dif12 = dif12 * (-1)
    
    tot_dist += dist_elevator * dif12
    
    list_sequence.append(floor2)
    floor1 = floor2 
    tot_travel -= 1

print("")
print(f"{list_sequence}")
print("")
print(f"The total distance elevator made is = {tot_dist}")
```
