# Maiores Notas lidas

```python
i = 1
max1 = 0
max2 = -1
while True:
    read = int(input("How many grades do you want to read: "))
    if read <= 0:
        print("The grade entered is less or igual to zero, Please enter a positive valor!\n")
    else:
        break


while (read > 0):
    num = int(input(f"enter the {i}º grade: "))
    
    if num > max1:
        if num > max2:
            if max1 > max2:
                max2 = max1
                max1 = num
            else:
                max2 = max2
                max1 = num
        else:
            max1 = max2
            max2 = num
    else:
        if num > max2:
            max1 = max1
            max2 = num
        else:
            max1 = max1 
            max2 = max2
    
    i +=1
    read -=1


print(f"\nThe two big grade are\nFirst max: {max1}\nSecond Max : {max2}")
```
