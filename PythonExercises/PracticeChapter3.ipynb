```python
words = ["apple", "bat", "bar", "atom", "book"]
by_letter = {}

for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)

by_letter
```

```python
#the code in [73] can be written like bellow
by_letters = {}
for word in words:
    letter = word[0]
    by_letters.setdefault(letter,[]).append(word)

by_letters
```

```python
##to do the same as above, we can use colletions 

from collections import defaultdict

by_letter = defaultdict(list)

for word in words:
    by_letter[word[0]].append(word)

by_letter
```

### Nested list comprehensions

Suppose we wanted to get a single list containing all names with two or more a’s in them. We could certainly do this with a simple for loop

```python
all_data = [["John", "Emily", "Michael", "Mary", "Steven"],["Maria","Abel Chimbua Wanda", "Juan", "Javier", "Natalia", "Pilar"]]
name_of_interest = []

for names in all_data:
    enough_as = [name for name in names if name.count("a") >= 2]
    name_of_interest.extend(enough_as)

name_of_interest
```

### 
You can actually wrap this whole operation up in a single nested list comprehension,
which will look like:

```python
result = [name for names in all_data for name in names if name.count("a") >= 2]

result
```

```python
states = ["   Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda", "south   carolina##", "West virginia?"]

import re

def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub("[?£!£#]","", value)
        value = value.title()
        result.append(value)
    return result

clean_strings(states)
```

```python
#or
def remove_punctuation(value):
    value = re.sub("[#?!@$%^]", "", value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_string(strings, ops):
    resut = []
    for value in strings:
        for func in ops:
            value = func(value)
        resut.append(value)
    return resut

clean_string(states, clean_ops)
```

###  Generators

A generator is a convenient way, similar to writing a normal function, to construct a
new iterable object. Whereas normal functions execute and return a single result at
a time, generators can return a sequence of multiple values by pausing and resuming
execution each time the generator is used.

To create a generator, use the yield keyword instead of return in a function:

```python
def squares(n=10):
    print(f"Gnerates squares from 1 to {n**2}")
    for i in range (1, n+1):
        yield i**2 

gen = squares()
gen
```

```python
for x in gen:
    print(x, end = " ")
```

###  Generator expressions

Another way to make a generator is by using a generator expression. This is a genera
tor analogue to list, dictionary, and set comprehensions. To create one, enclose what
would otherwise be a list comprehension within parentheses instead of brackets:

```python
gen = (x**2 for x in range(100))
gen
```

```python
# This is equivalent to the following more verbose generator:

def _make_gen():
    for x in range(100):
        yield x**2

gen = _make_gen()
gen 
```

```python
"""Generator expressions can be used instead of list comprehensions as function arguments in some cases"""

sum(x**2 for x  in range(100))
```

```python
"""Depending on the number of elements produced by the comprehension expression,the generator version can sometimes be meaningfully faster."""

dict((i,i**2) for i in range(5))
```
