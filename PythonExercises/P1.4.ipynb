### Difference age according to birthday

```python
print("Read Person's one birthday")
dd1 = int(input("Day: "))
mm1 = int(input("Mot: "))
aa1 = int(input("Yea: "))

print("Read Person's two birthday")
dd2 = int(input("Day: "))
mm2 = int(input("Mot: "))
aa2 = int(input("Yea: "))

if dd1 > dd2:
    if mm1 > mm2:
        if aa1 > aa2:
            print("PERSON TWO IS YOUGEST, by DD/MM/YY")
        else:
            print("PERSON TWO IS YOUGEST by DD and MM, but SMALLEST by YY")
    else:
        if aa1 > aa2:
            print("PERSON TWO IS YOUGEST by DD and YY, but SMALLEST by MM")
        else:
            print("PERSON TWO IS YOUGEST by DD, but SMALLEST by MM and YY")
else:
    if mm1 > mm2:
        if aa1 > aa2:
            print("PERSON ONE IS YOUGEST by DD, but SMALLEST by MM and YY")
        else:
            print("PERSON ONE IS YOUGEST by DD and YY, but SMALLEST by MM")
    else:
        if aa1 > aa2:
            print("PERSON ONE IS YOUGEST by DD and MM, but SMALLEST by YY")
        else:
            print("PERSON ONE IS YOUGEST by DD/MM/YY")
```
